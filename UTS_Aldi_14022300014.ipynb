```python
from google_play_scraper import reviews, Sort
import csv

result, _ = reviews(
    'com.michatapp.im',
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=100,
    filter_score_with=None
)

filename = 'ulasan_google_play.csv'


with open(filename, 'w', newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=['userName', 'score', 'at', 'content'])
    writer.writeheader()
    for review in result:

        writer.writerow({
            'userName': review['userName'],
            'score': review['score'],
            'at': review['at'],
            'content': review['content']
        })

print(f"Berhasil menyimpan {len(result)} ulasan ke '{filename}'")
```

```python
import pandas as pd
from transformers import pipeline

# Load the sentiment analysis pipeline
sentiment_pipeline = pipeline("sentiment-analysis", model="w11wo/indonesian-roberta-base-prdect-id")
```

```python
# Load the CSV file into a DataFrame
df_reviews = pd.read_csv(filename)

# Perform sentiment analysis on the 'content' column
# The model expects a list of strings
sentiment_results = sentiment_pipeline(df_reviews['content'].tolist())

# Extract labels and scores
df_reviews['sentiment'] = [res['label'] for res in sentiment_results]
df_reviews['sentiment_score'] = [res['score'] for res in sentiment_results]

# Display the DataFrame with sentiment analysis results
display(df_reviews.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the occurrences of each sentiment label
sentiment_counts = df_reviews['sentiment'].value_counts().reset_index()
sentiment_counts.columns = ['Sentiment', 'Count']

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Sentiment', y='Count', data=sentiment_counts, palette='viridis', hue='Sentiment', legend=False)
plt.title('Distribusi Sentimen Ulasan Pengguna')
plt.xlabel('Sentimen')
plt.ylabel('Jumlah Ulasan')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Grafik di atas menunjukkan distribusi sentimen dari ulasan-ulasan yang telah dianalisis. Anda dapat melihat proporsi ulasan yang dikategorikan sebagai 'Happy', 'Anger', 'Fear', 'Sadness', 'Love', dan sentimen lainnya.
